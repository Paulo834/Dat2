```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/antropometria-dataset-1.csv')
df.head()
```

```python
sns.boxplot(data = df.loc[:,
                          ['peso', 'cintura', 'cadera']
                         ]
           )


plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Box plots', fontsize = 17)
plt.show()   
```

```python
sns.boxplot(data = df.loc[:,
                          ['sistol', 'diastol']
                         ]
           )


plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Box plots', fontsize = 17)
plt.show()   
```

```python
df.plot(kind = 'line')

plt.xticks( rotation = 45, fontsize = 17)
plt.yticks(rotation = 45, fontsize = 17)
plt.xlabel('x_label', fontsize = 17)
plt.ylabel('y_name', fontsize = 17)

plt.title('Line plot', fontsize = 17)

plt.show()
```
