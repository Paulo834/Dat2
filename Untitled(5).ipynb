# The higher the ranking, the higher the TDs? 

```python
sns.lmplot(x = 'Off Rank',
           y = 'Off TDs',
           data = df,
           fit_reg = False,
           
          )
plt.title('Official Rank vs Touchdowns', fontsize = 19)
plt.show()
```

Conclusion

We can see that the hypothesis is true, if a team has a better rank it will have more touchdowns.
